```python
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
np.random.seed(0)
X = np.random.uniform(low=-1, high=+1, size=(5000, 100, 100))
mean_of_median = np.mean(np.median(X, axis=2), axis=1)
print(np.mean(mean_of_median), np.std(mean_of_median))
median_of_mean = np.median(np.mean(X, axis=2), axis=1)
print(np.mean(median_of_mean), np.std(median_of_mean))
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(mean_of_median, label='Mean of Medians', color='blue', kde=True)
sns.histplot(median_of_mean, label='Median of Means', color='red', kde=True)
plt.legend()

from scipy.stats import kstest
ks_stat, ks_p_value = kstest(mean_of_median, median_of_mean)

print(f'K-S Test Statistic: {ks_stat}')
print(f'K-S Test P-value: {ks_p_value}')
```
